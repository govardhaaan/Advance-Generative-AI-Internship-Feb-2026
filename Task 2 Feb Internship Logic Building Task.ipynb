```python
sentence = input("Enter sentence:")
words = sentence.split()
unique_words = set(words)
print("Unique words count:", len(unique_words))
print("Unique words:", unique_words)

```

```python
employees = {
    "Govardhan": 70000,
    "Sai": 68000,
    "Badri": 72000,
    "abhi": 65000
}
employee = max(employees, key=employees.get)
highest_salary = employees[employee]
print("Highest Salary:", employee, "-", highest_salary)
```

```python
numbers = [55, 14, 80, 99, 10]
maximum = numbers[0]
minimum = numbers[0]
for num in numbers:
    if num > maximum:
        maximum = num
    if num < minimum:
        minimum = num
print("List:", numbers)
print("Max:", maximum)
print("Min:", minimum)
```

```python
prices = [550, 1400, 999, 1800, 500]
count = 0
for price in prices:
    if price > 1000:
        count += 1
print("Products above 1000:", count)
```

```python
attendance = ["P", "P", "A", "P", "P","A"]
total_days = len(attendance)
present_days = 0
for day in attendance:
    if day == "P":
        present_days += 1
percentage = (present_days / total_days) * 100
print("Attendance Percentage:", percentage)
```

```python
text = "maneesha"
char_count = {}
for ch in text:
    if ch in char_count:
        char_count[ch] += 1
    else:
        char_count[ch] = 1
print(char_count)
```
