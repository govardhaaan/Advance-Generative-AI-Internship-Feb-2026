```python
def validate_recharge(amount):
    valid_plans = [199, 299, 399, 599]
    if amount >= 50 and amount in valid_plans:
        return True
    else:
        return False
def start_recharge_system():
    print("--- Welcome to the Mobile Recharge Portal ---")

    while True:
        try:
            user_input = int(input("\nEnter recharge amount (e.g., 199, 299, 399, 599): "))

            if validate_recharge(user_input):
                print(f" Success! Your recharge of ₹{user_input} is processing.")
                break
            else:
                print(" Invalid Amount. Please ensure the amount is ≥ ₹50 and is a valid plan.")

        except ValueError:
            print("⚠️ Please enter a numeric value only.")
start_recharge_system()
```

```python
def check_inventory(inventory_data):
    threshold = 15
    print(f"{'Product':<15} | {'Status':<15}")
    print("-" * 35)

    for product, stock in inventory_data.items():
        if stock < threshold:
            print(f"{product:<15} |  Reorder Alert ({stock})")
        else:
            print(f"{product:<15} |  Stock OK ({stock})")
warehouse_stock = {
    "Apples": 50,
    "Bananas": 12,
    "Milk": 8,
    "Bread": 25,
    "Eggs": 10
}
check_inventory(warehouse_stock)
```

```python
def process_result(marks):
    total = 0
    for mark in marks:
        total += mark
    average = total / len(marks)

    print("Average Marks:", average)
    if average >= 50:
        print("Result: Pass")
        return "Pass"
    else:
        print("Result: Fail")
        return "Fail"
student_marks = [60, 45, 70, 55]
process_result(student_marks)
```

```python
def calculate_fare(distance, peak):
    base_fare = 50
    rate_per_km = 12
    total_fare = base_fare + (distance * rate_per_km)
    if peak.lower() == "yes":
        total_fare += total_fare * 0.25

    return total_fare
while True:
    distance = float(input("Enter distance in km: "))
    peak = input("Is it peak hour? (yes/no): ")

    fare = calculate_fare(distance, peak)
    print("Total Fare: ₹", fare)

    retry = input("Do you want to calculate again? (yes/no): ")

    if retry.lower() != "yes":
        print("Thank you for using Cab Fare Estimator")
        break
```

```python
def check_eligibility(attendance):
    present_days = 0
    for day in attendance:
        if day == "P":
            present_days += 1

    total_days = len(attendance)
    if total_days == 0:
        return "No Attendance Data"

    percentage = (present_days / total_days) * 100
    print("Attendance Percentage:", percentage)

    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"
attendance_list = ["P", "A", "P", "P", "A", "P", "P", "P"]
result = check_eligibility(attendance_list)
print("Eligibility Status:", result)
```

```python
def check_password_strength(password):
    has_digit = False
    has_special = False
    special_chars = "@#$"
    if len(password) < 8:
        return "Weak Password (Minimum 8 characters required)"
    for char in password:
        if char.isdigit():
            has_digit = True
        if char in special_chars:
            has_special = True
    if has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password (Missing digit or special character)"
pwd = input("Enter your password: ")
result = check_password_strength(pwd)
print(result)
```
