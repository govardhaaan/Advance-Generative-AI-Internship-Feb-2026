```python
#2
marks={45,78,90,33,60}
pass_count=0
fail_count=0
for score in marks:
  if score >= 50:
    pass_count +=1
  else:
    fail_count +=1
print(f"total pass students:{pass_count}")
print(f"total fail student:{fail_count}")
```

```python
#4
messages = ["Hi", "Welcome to the platform", "OK"]
for msg in messages:
  length = len(msg)
  print("Message:", msg)
  print("Length:", length)
  if length > 10:
    print("Status: Longer than 10 characters ")
  else:
    print("Status: Within limit ")
    print()
```
