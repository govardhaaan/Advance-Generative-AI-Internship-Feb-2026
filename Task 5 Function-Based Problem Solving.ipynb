```python
def analyze_engagement(likes_list):
    total_likes = 0
    for likes in likes_list:
        total_likes += likes
    print("Total Likes:", total_likes)
    if total_likes >= 1000:
        print("Post Status: Viral Post")
    else:
        print("Post Status: Normal Engagement")
likes = [200, 80, 300, 400]
analyze_engagement(likes)
```

```python
def monitor_stock(stock):
    print("Medicine Stock:", stock)
    if stock < 10:
        print("Status: Low Stock Alert")
    else:
        print("Status: Stock Sufficient")
monitor_stock(60)
```

```python
def analyze_rainfall(rainfall_data, required_level):
    total = 0
    for rainfall in rainfall_data:
        total += rainfall
    average = total / len(rainfall_data)

    print("Average Rainfall:", int(average))
    if average >= required_level:
        print("Rainfall Status: Adequate Rainfall")
    else:
        print("Rainfall Status: Inadequate Rainfall")
rainfall = [80, 89, 95, 83]
required = 70
analyze_rainfall(rainfall, required)
```

```python
def detect_duplicate_usernames(usernames):
    unique_users = set()
    duplicates_found = False
    for user in usernames:
        if user in unique_users:
            duplicates_found = True
            break
        unique_users.add(user)
    if duplicates_found:
        print("Duplicate Accounts Found: Yes")
    else:
        print("Duplicate Accounts Found: No")
users = ["user9", "user2", "users3", "user1"]
detect_duplicate_usernames(users)
```

```python
def filter_premium_crops(crop_prices):
    premium_threshold = 2000
    premium_crops = []
    for price in crop_prices:
        if price > premium_threshold:
            premium_crops.append(price)
    for price in crop_prices:
        if price > premium_threshold:
            print(f"₹{price} → Premium Crop")
        else:
            print(f"₹{price} → Regular Crop")
prices = [1500, 2500, 1800, 3000]
filter_premium_crops(prices)
```

```python
def evaluate_application_health():
    error_count = int(input("Enter Error Count: "))
    print("Error Count:", error_count)
    if error_count == 0:
        print("Application Status: Healthy")
    elif error_count <= 5:
        print("Application Status: Minor Issues")
    else:
        print("Application Status: Critical Issues")
evaluate_application_health()
```

```python
def check_exam_eligibility(attendance_list):

    total_days = len(attendance_list)
    present_days = 0
    for status in attendance_list:
        if status == "P":
            present_days += 1
    percentage = (present_days / total_days)
    print("Attendance Percentage:", round(percentage, 2), "%")
    if percentage >= 75:
        print("Exam Eligibility: Eligible")
    else:
        print("Exam Eligibility: Not Eligible")
attendance = ["P", "A", "P", "P", "A", "P", "P", "P"]
check_exam_eligibility(attendance)
```

```python
def check_exam_eligibility():
    total_days = int(input("Enter total number of classes: "))
    present_days = 0
    for i in range(total_days):
        status = input(f"Enter attendance for day {i+1} (P/A): ").upper()
        if status == "P":
            present_days += 1
    percentage = (present_days / total_days) * 100
    print("Attendance Percentage:", round(percentage, 2), "%")
    if percentage >= 75:
        print("Exam Eligibility: Eligible")
    else:
        print("Exam Eligibility: Not Eligible")
check_exam_eligibility()
```
