```python
import string
Query = "govardhan buy phone use phone"
Query = Query.lower()
Query = Query.translate(str.maketrans('', '', string.punctuation))
words = Query.split()
word_count = {}
for word in words:
    word_count[word] = word_count.get(word, 0) + 1
repeated_words = {word: count for word, count in word_count.items() if count > 1}
print(repeated_words)
```

```python
s_readings = [8,2,7,9,6,4,10]
v_readings = []
for hour, value in enumerate(s_readings):
    if value % 2 == 0:
        v_readings.append((hour, value))
print("Valid Sensor Readings (Hour, Value):")
print(v_readings)
```

```python
emails = [
    "Govardhan@gmail.com",
    "Sai@yahoo.com",
    "Abhi@gmail.com",
    "Umanth@gmail.com",
    "Vyshu@yahoo.com"
]
domain_count = {}
for email in emails:
    domain = email.split('@')[1]
    domain_count[domain] = domain_count.get(domain, 0) + 1
total_emails = len(emails)
print("Email Domain Usage Percentage:")
for domain, count in domain_count.items():
    percentage = (count / total_emails) * 100
    print(f"{domain}: {int(percentage)}%")
```

```python
sales = [1100, 1600, 800, 2500, 1900, 5000]
average_sales = sum(sales) / len(sales)
threshold = average_sales * 1.30
print("Sales Spikes Detected:")
for day, value in enumerate(sales, start=1):
    if value > threshold:
        print(f"Day {day}: {value}")
```

```python
from collections import Counter
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]
id_count = Counter(user_ids)
print("Duplicate User IDs:")
for user_id, count in id_count.items():
    if count > 1:
        print(f"{user_id} → {count} times")
```
