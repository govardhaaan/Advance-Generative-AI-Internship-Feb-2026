```python
def smart_parking_system(capacity, vehicle_logs):
    parked = 0
    peak = 0

    for log in vehicle_logs:
        if log == "IN":
            parked += 1
        elif log == "OUT":
            if parked > 0:
                parked -= 1

        if parked > peak:
            peak = parked

    print("Currently Parked Vehicles:", parked)
    print("Peak Usage:", peak)

    if parked > capacity:
        print("Parking Status: Over Capacity Alert!")
    else:
        print("Parking Status: Available")
smart_parking_system(50, ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"])
```

```python
def delivery_time_estimator(distance, traffic, weather):
    base_time = distance * 5
    if traffic == "High":
        base_time += 15
    elif traffic == "Medium":
        base_time += 10
    if weather == "Rainy":
        base_time += 10
    elif weather == "Storm":
        base_time += 20
    print("Estimated Delivery Time:", base_time, "minutes")
delivery_time_estimator(8, "High", "Rainy")
```

```python
def theatre_occupancy(total_seats, booked_seats):
    booked_count = len(booked_seats)
    occupancy = (booked_count / total_seats) * 100
    print("Occupancy:", round(occupancy), "%")
    if occupancy == 100:
        print("Show Status: Housefull")
    elif occupancy >= 75:
        print("Show Status: Almost Full")
        print("Suggestion: Consider Opening Additional Show")
    else:
        print("Show Status: Seats Available")
theatre_occupancy(200, [1]*150)
```

```python
def server_load_classification(cpu_readings):
    total = 0
    for reading in cpu_readings:
        total += reading
    avg = total / len(cpu_readings)
    print("Average CPU Load:", round(avg), "%")
    if avg < 50:
        print("Server Status: Normal")
    elif avg <= 80:
        print("Server Status: Warning")
    else:
        print("Server Status: Critical")
server_load_classification([45, 60, 70, 85, 90])
```

```python
def classroom_resource_monitor(resources):
    overused = []
    for resource, hours in resources.items():
        if hours > 8:
            overused.append(resource)
    print("Overused Resources:", ", ".join(overused))
    if overused:
        print("Energy Alert: Yes")
    else:
        print("Energy Alert: No")
resource_data = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}

classroom_resource_monitor(resource_data)
```

```python
def event_registration_controller(capacity, registrations):
    if registrations > capacity:
        confirmed = capacity
        waitlist = registrations - capacity
        status = "Closed"
    else:
        confirmed = registrations
        waitlist = 0
        status = "Open"
    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlist)
    print("Registration Status:", status)
event_registration_controller(100, 105)
```
